```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Define mean and covariance matrix for the multivariate normal distribution
mean = np.array([0, 0])
covariance = np.array([[1, 0.5], [0.5, 1]])

# Create a grid of points in 2D
x, y = np.mgrid[-5:5:0.1, -5:5:0.1]
pos = np.dstack((x, y))

# Create the multivariate normal distribution
rv = multivariate_normal(mean, covariance)

# Calculate the probability density function (PDF) for each point in the grid
pdf = rv.pdf(pos)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the 3D Gaussian distribution
ax.plot_surface(x, y, pdf, cmap='viridis')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('d')

# Show the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Parameters for the multivariate normal distribution
mean_normal = np.array([0, 0])
covariance_normal = np.array([[1, 0.5], [0.5, 1]])

# Parameters for the circular normal distribution
center_circle = np.array([0, 0])
radius_circle = 10

# Create a grid of points in 2D
x, y = np.mgrid[-15:15:0.1, -15:15:0.1]
pos = np.dstack((x, y))

# Create the multivariate normal distribution
rv_normal = multivariate_normal(mean_normal, covariance_normal)

# Calculate the probability density function (PDF) for each point in the grid
pdf_normal = rv_normal.pdf(pos)

# Create a circular mask
mask_circle = np.sqrt((x - center_circle[0])**2 + (y - center_circle[1])**2) <= radius_circle

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the circular normal distribution
ax.plot_surface(x, y, pdf_normal, cmap='viridis', alpha=0.5)

# Plot the circular region
ax.plot_surface(x * mask_circle, y * mask_circle, np.zeros_like(pdf_normal), color='red', alpha=0.2)

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('d')

# Show the plot
plt.show()
```
